```python
url='https://www.alphavantage.co/query?function=DIGITAL_CURRENCY_WEEKLY&symbol=BTC&market=USD&apikey='+str(API_key)
r= requests.get(url)
data= r.json()
```

```python
print(data)
```

```python
ts = data['Time Series (Digital Currency Weekly)']
```

```python
print(ts)
```

```python
df = pd.DataFrame(columns=['Date', 'Open', 'High', 'Low', 'Close', 'Volume'])
```

```python
for date, values in ts.items():
    df = df.append({'Date': date, 'Open': values['1b. open (USD)'], 'High': values['2b. high (USD)'],
                    'Low': values['3b. low (USD)'], 'Close': values['4b. close (USD)'], 'Volume': values['5. volume']},
                   ignore_index=True)
```

```python
df.columns
```

```python
df.head()
```

```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.set_index('Date',inplace=True)
```

```python
df['Open']
```

```python

import matplotlib.pyplot as plt


```

```python
# Plot 'Open' column
df['Open'].plot()
plt.xlabel('Date')
plt.ylabel('Open')
plt.title('BTC Open Prices Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
columns = ['Open', 'High', 'Low', 'Close', 'Volume']
df[columns] = df[columns].apply(pd.to_numeric, errors='coerce')
```

```python
df.head()
```

```python
df.plot()
```

```python
enumerate(df.columns)
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(len(df.columns), 1, figsize=(10, 6*len(df.columns)))

for i, column in enumerate(df.columns):
    axs[i].plot(df.index, df[column])
    axs[i].set_xlabel('Index')
    axs[i].set_ylabel(column)
    axs[i].set_title(f'{column} Variation')

plt.tight_layout()
plt.show()
```
